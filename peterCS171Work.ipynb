```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Define regions
regions = {
    "Northeast": ["Connecticut", "Maine", "Massachusetts", "New Hampshire", "Rhode Island", "Vermont", "New Jersey", "New York", "Pennsylvania"],
    "Midwest": ["Illinois", "Indiana", "Michigan", "Ohio", "Wisconsin", "Iowa", "Kansas", "Minnesota", "Missouri", "Nebraska", "North Dakota", "South Dakota"],
    "South": ["Delaware", "Florida", "Georgia", "Maryland", "North Carolina", "South Carolina", "Virginia", "District of Columbia", "West Virginia", "Alabama", "Kentucky", "Mississippi", "Tennessee", "Arkansas", "Louisiana", "Oklahoma", "Texas"],
    "West": ["Arizona", "Colorado", "Idaho", "Montana", "Nevada", "New Mexico", "Utah", "Wyoming", "Alaska", "California", "Hawaii", "Oregon", "Washington"]
}

# Function to perform PCA for a given region and display the required information
def perform_pca_for_region(region_name, region_states, data):
    # Filter data for the region
    region_data = data[data['State'].isin(region_states)]
    
    # Select the features for PCA
    features = region_data.drop(columns=['Year', 'State', 'fertility_rate'])
    
    # Standardize the features
    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(features)
    
    # Perform PCA
    pca = PCA()
    pca.fit(scaled_features)
    
    # Get the explained variance ratio
    explained_variance = pca.explained_variance_ratio_
    
    # Get the PCA components
    components = pca.components_
    
    # Create a DataFrame to display the impact of each variable
    impact_df = pd.DataFrame(components, columns=features.columns)
    impact_df.index = [f'PC{i+1}' for i in range(len(impact_df))]
    
    # Calculate cumulative explained variance
    cumulative_explained_variance = np.cumsum(explained_variance)
    
    # Display cumulative explained variance
    print(f"Cumulative Explained Variance for {region_name}:")
    print(cumulative_explained_variance)
    
    # Display the impact of each variable on the principal components
    print(f"\nImpact of Each Variable on Principal Components for {region_name}:")
    print(impact_df)
    
    # Display the first 6 principal components
    print(f"\nFirst 6 Principal Components for {region_name}:")
    print(components[:6])
    
    # Sum the first 6 principal components
    sum_first_6_pcs = np.sum(components[:6], axis=0)
    
    # Display the sum of the first 6 principal components
    print(f"\nSum of the First 6 Principal Components for {region_name}:")
    print(sum_first_6_pcs)
    
    # Get the indices of the top 5 values of the sum_first_6_pcs
    top_5_indices = np.argsort(sum_first_6_pcs)[-5:]
    
    # Display the indices of the top 5 values
    print(f"\nIndices of the Top 5 Values of the Sum of the First 6 Principal Components for {region_name}:")
    print(top_5_indices)
    
    # Get the variable names corresponding to the top 5 indices
    top_5_variables = features.columns[top_5_indices]
    
    # Display the top 5 variables
    print(f"\nTop 5 Variables Corresponding to the Sum of the First 6 Principal Components for {region_name}:")
    print(top_5_variables)
    print("\n" + "="*50 + "\n")

# Load your data
# data = pd.read_csv('your_data.csv')  # Replace with your data loading method

# Perform PCA for each region
for region_name, region_states in regions.items():
    perform_pca_for_region(region_name, region_states, data)
```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Define regions
regions = {
    "Northeast": ["Connecticut", "Maine", "Massachusetts", "New Hampshire", "Rhode Island", "Vermont", "New Jersey", "New York", "Pennsylvania"],
    "Midwest": ["Illinois", "Indiana", "Michigan", "Ohio", "Wisconsin", "Iowa", "Kansas", "Minnesota", "Missouri", "Nebraska", "North Dakota", "South Dakota"],
    "South": ["Delaware", "Florida", "Georgia", "Maryland", "North Carolina", "South Carolina", "Virginia", "District of Columbia", "West Virginia", "Alabama", "Kentucky", "Mississippi", "Tennessee", "Arkansas", "Louisiana", "Oklahoma", "Texas"],
    "West": ["Arizona", "Colorado", "Idaho", "Montana", "Nevada", "New Mexico", "Utah", "Wyoming", "Alaska", "California", "Hawaii", "Oregon", "Washington"]
}

# Dictionary to store the top 5 variables and their values for each region
top_5_variables_per_region = {}

for region_name, region_states in regions.items():
    # Filter data for the region
    region_data = data[data['State'].isin(region_states)]
    
    # Select the features for PCA
    features = region_data.drop(columns=['Year', 'State', 'fertility_rate'])
    
    # Standardize the features
    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(features)
    
    # Perform PCA
    pca = PCA()
    pca.fit(scaled_features)
    
    # Get the PCA components
    components = pca.components_
    
    # Sum the first 6 principal components
    sum_first_6_pcs = np.sum(components[:6], axis=0)
    
    # Get the indices of the top 5 values of the sum_first_6_pcs
    top_5_indices = np.argsort(sum_first_6_pcs)[-5:]
    
    # Get the variable names and their corresponding values
    top_5_variables = features.columns[top_5_indices]
    top_5_values = sum_first_6_pcs[top_5_indices]
    
    # Store the top 5 variables and their values for the region
    top_5_variables_per_region[region_name] = {
        'variables': top_5_variables.tolist(),
        'values': top_5_values.tolist()
    }

# Display the top 5 variables and their values for each region
for region_name, top_5_data in top_5_variables_per_region.items():
    print(f"Top 5 Variables for {region_name}:")
    print(top_5_data['variables'])
    print(f"Values: {top_5_data['values']}")
    print("\n" + "="*50 + "\n")

# Create a DataFrame from the dictionary
top_5_df = pd.DataFrame.from_dict(top_5_variables_per_region, orient='index')

# Reset the index to have region names as a column
top_5_df.reset_index(inplace=True)
top_5_df.columns = ['Region', 'Variables', 'Values']

# Save the DataFrame to a CSV file
top_5_df.to_csv('data/top_5_variables_per_region.csv', index=False)
```
